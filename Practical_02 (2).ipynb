```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.svm import SVC
```

```python
df = pd.read_csv(r"C:\Users\kartikeshwar singh\Downloads\archive (8)\emails.csv")
```

```python
df.head()
```

```python
df.columns
```

```python
df.isnull().sum()
```

```python
df.drop(['Email No.'], axis = 1,inplace = True)
```

```python
X = df.drop(['Prediction'],axis = 1)
y = df['Prediction']
```

```python
from sklearn.preprocessing import scale
X = scale(X)
X_train,X_test,y_train,y_test = train_test_split(X, y, test_size = 0.3,random_state = 43)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 7)
knn.fit(X_train, y_train)
```

```python
y_pred = knn.predict(X_test)
print("Predicted values:\n", y_pred)
```

```python
print("Predicted values:", y_pred)
```

```python
print("Confusion Matrix:\n",metrics.confusion_matrix(y_test, y_pred))
```

```python
print("Accuracy Score:\n",metrics.accuracy_score(y_test, y_pred) * 100)
```

```python
model = SVC(C = 1)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
print("Predicted values:\n", y_pred)
```

```python
print("Accuray Score of SVM:",metrics.accuracy_score(y_test, y_pred) * 100)
```

```python
print("Confusion Matrix:\n",metrics.confusion_matrix(y_true = y_test, y_pred = y_pred))
```
