```python
# Gradient Descent to find the local minima of y = (x + 3)^2

# Define the function and its derivative
def function(x):
    return (x + 3) ** 2

def derivative(x):
    return 2 * (x + 3)

# Gradient Descent parameters
x = 2                    # Starting point
learning_rate = 0.1      # Step size
precision = 1e-6         # Precision for stopping criteria
max_iters = 1000         # Maximum number of iterations

# Gradient Descent Loop
for i in range(max_iters):
    # Calculate the gradient
    grad = derivative(x)
    
    # Update x by moving in the direction of the negative gradient
    new_x = x - learning_rate * grad
    
    # Check if the change is within the precision threshold
    if abs(new_x - x) < precision:
        break
    
    x = new_x  # Update x

# Results
print("The local minima occurs at x =", x)
print("The function value at local minima f(x) =", function(x))
```

```python
import numpy as np
import matplotlib.pyplot as plt

def function(x):
    return (x + 3) ** 2

def derivative(x):
    return 2 * (x + 3)

x = 2
learning_rate = 0.1
precision = 1e-6
max_iters = 1000

x_values = [x]
f_values = [function(x)]

for i in range(max_iters):
    gradient = derivative(x)
    x_new = x - learning_rate * gradient

    x_values.append(x_new)
    f_values.append(function(x_new))

    if abs(x_new - x) < precision:
        break

    x = x_new

x_range = np.linspace(-6, 4, 100)
y_range = function(x_range)

plt.figure(figsize=(10, 6))
plt.plot(x_range, y_range, label='f(x) = (x + 3) ** 2)',color = 'blue')
plt.scatter(x_values, f_values,color = 'red',marker = 'o')

plt.scatter(x_values[-1], f_values[-1], color = 'green',s = 100)

print("Local minima occurs at x =", x)
print("The function value at Local minima f(x)", function(x))
```
